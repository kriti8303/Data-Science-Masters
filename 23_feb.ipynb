## Assignment Question 4:

DataFrame : It is a two-dimensional, size-mutable, potentially hetrogenous tabular data structure with labeled axes(rows and columns).

Difference between dataframe and series : 
A pandas series is a one-dimensional data structure that comprises of key-value pair in which element must be of same data type.

A pandas dataframe is a multi-dimensional data structure that comprises key value pair in which element must be of different data types.

```python
## example of pandas series
pd.Series([1, 7, 2])
```

```python
## example of pandas dataframe
pd.DataFrame({
    "calories" : [420, 380, 390],
    "duration" : [50, 40, 45]
})
```

## Assignment Question 7:

You can create a DataFrame from multiple Series objects by adding each series as a columns. By using concat() method you can merge multiple series together into dataframe. This takes several params, for our scenario we use list that takes series to combine and axis=1 to specify merge series as columns instead of rows. 

```python
## example 
import pandas as pd
```

```python
author = ['Jitendra', 'purnima', 'arpit', 'jyoti']
```

```python
article = [210, 211, 114, 178]
```

```python
auth_series = pd.Series(author)
```

```python
article_series = pd.Series(article)
```

```python
frame = {'Author' : auth_series, 'Article' : article_series }
```

```python
result = pd.DataFrame(frame)
```

```python
print(result)
```
